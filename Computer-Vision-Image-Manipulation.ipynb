```python
import os
from PIL import Image
```

```python
### Open image and convert to grayscale.

# Place image file directory in the same location as Jupyter notebook file if any issues locating file
file_dir = "led"
file_name = "10.png" #name of the image

# Use the os library to create a full path (regardless of operating system used)
path = os.path.join(file_dir, file_name)

img = None
img = Image.open(path)


# Convert the image to grayscale
img = img.convert('L')

#To preview image use img.rotate(45).show() in PIL library
```

```python
#Also can preview images using imshow() function in the matplotlib library.
import matplotlib.pyplot as plt
```

```python
### View the image

plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
#Convert Image to Numpy Array

import numpy as np
```

```python
# Convert the Image object to a Numpy array. Python doesn't have built-in arrays 

img_array = None

img_array =np.asarray(img)
```

```python
# Show array as image
plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
```

```python
#Details about the array:

#Shape of image array
print(np.shape(img_array))
```

```python
# First rows of image array
print(img_array[0,:])
```

```python
# First column of image array
print(img_array[:,0])
```

```python
# A method to print the full array in Python:
np.set_printoptions(threshold=np.inf)
print(img_array)
```

```python
#The Scikit-Image Transform module will handle resizing the image
from skimage.transform import resize
```

```python
# Resize image array to 28x28 pixels.
# The "anti_aliasing" parameter set to 'True' in order smooth out the 'hard'edges in image

img_resized = None
img_resized = resize(img_array,(28,28),anti_aliasing=True)
print(np.shape(img_resized))
```

```python
# View the new array as an image. Initially image will appear black 

plt.imshow(img_resized, cmap='gray', vmin=0, vmax=255)
```

```python
# Resize function in skimage automatically normalizes all of the values in the image to be between 0.0 and 1.0 
# Multiply a Numpy array by a constant 255
img_resized = img_resized * 255
```

```python
#Round array values to integers
img_resized = np.rint(img_resized)

img_resized=img_resized.astype(int)
print(img_resized)

# Clamp values between 0 and 255
img_resized = np.clip(img_resized,0,255)
```
